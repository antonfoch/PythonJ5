# Exercice 4

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('D-POO-300_05_dataset-stock.csv')


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))


data.plot(x='date', y='close', ax=axes[0,0], title='Plot 1')
data.plot(x='date', y='close', ax=axes[0,1], title='Plot 2')
data.plot(x='date', y='close', ax=axes[1,0], title='Plot 3')
data.plot(x='date', y='close', ax=axes[1,1], title='Plot 4')

plt.tight_layout()
plt.show()
```

# Exercice 4 bis 

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('D-POO-300_05_dataset-stock.csv')


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))


data.plot(x='date', y='close', ax=axes[0,0], title='A')
data.plot(x='date', y='close', ax=axes[0,1], title='B', linewidth=3, color='green')
data.plot(x='date', y='close', ax=axes[1,0], title='C', color = 'purple',  marker='o', markersize=3)
data.plot(x='date', y='close', ax=axes[1,1], title='D', marker='s', markersize=2, markeredgecolor="red" , color='black', linestyle='-', linewidth=1,)


plt.tight_layout()
plt.show()
```

# Exercise 05

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('D-POO-300_05_dataset-stock.csv')


data['date'] = pd.to_datetime(data['date'])

plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['open'], label='Series 1', color='blue')
plt.plot(data['date'], data['high'], label='Series 2', color='red')
plt.plot(data['date'], data['low'], label='Series 3', color='green')
plt.plot(data['date'], data['close'], label='Series 4', color='purple')


plt.title('Four Series from stock.csv')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
